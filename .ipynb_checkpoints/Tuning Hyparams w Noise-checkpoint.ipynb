+ How much does hyparam chosen with noisy label affect ultimate testing performance?
+ What classifiers' evaluation curve are more robust i.e. consistent? What makes them consistent?
+ Does classifiers always choose high regularizer value to avoid overfitting?
+ For bagging and others, OOB vs cross-validation?

+ Mean & std of scores for each hyparam value under noise vs clean? Is std higher for noisy case? Is that why wrong value gets chosen? **If** high variance is the culprit, isn't increasing no of runs for *each hyparam value* make this more stable, and thus help best value to win?
+ Relation between estimated validation accuracy and general test error? What factors play a role?
+ What metrics are more consistent (less variance) under noise? Some works already tackle it for clean case. Ones that work on probability can be *proven* to work better. Do they also have closer estimate to true value?

```python
ds = ['camel-1.4.0.csv','groovy-1_6_BETA_1.csv','jruby-1.5.0.csv','lucene-2.9.0.csv']
X,yn,y = read_jira(ds[0])
lcm = confusion_matrix(y,yn,normalize='true')
lcm
```

## Impact of Noise on Hyparam-Performance curve

```python
HYP = 'C'
pars = {
    HYP:[.05,.1,.25,.5,1.0,2,4,10,25]
}
clf = SVC(kernel='poly')
cv5 = StratifiedShuffleSplit(n_splits=5,random_state=49)
CVr = RepeatedStratifiedKFold(n_repeats=3,n_splits=5,random_state=99)
SCORERS = [precision_score,recall_score,f1_score,accuracy_score,matthews_corrcoef,balanced_accuracy_score]
```

```python
def validate(clf,params,X,Ytrain,Yeval,CV,eval_metrics):
    global HYP,lcm
    res = defaultdict(list)
    for h in params[HYP]:
        est = clone(clf)
        est.set_params(**{HYP:h})
        r = noisy_evaluate(est,X,Ytrain,Yeval,CV,eval_metrics)
        for m in SCORERS:
            res[m].append(r[m])
    return res
```

```python
noisy = validate(clf,pars,X,yn,yn,cv5,SCORERS)
```

```python
clean = validate(clf,pars,X,yn,y,cv5,SCORERS)
```

```python
fig,axs = plt.subplots(nrows=2,ncols=3,figsize=(12,6))
fig.tight_layout()
axs = axs.ravel()
for it,m in enumerate(SCORERS):
    axs[it].plot(np.log(pars[HYP]),clean[m],label='Yclean');
    axs[it].plot(np.log(pars[HYP]),noisy[m],label='Ynoisy');
    axs[it].scatter(np.log(pars[HYP]),clean[m]);
    axs[it].scatter(np.log(pars[HYP]),noisy[m]);
    axs[it].set_title(m.__name__)
    axs[it].legend();
```

## Impact of Tuning Hyparam under Noise on Final Test Error

```python
def test_error(clf,params,X,Ytrain,Yeval,Ytest,CVval,CVtest,eval_metrics,target_metric):
    global HYP
    scores = defaultdict(list)
    for train_id,test_id in CVtest.split(X,Ytest):
        res = validate(clf,params,X[train_id],Ytrain[train_id],Yeval[train_id],CVval,eval_metrics)
        best_idx = np.argmax(res[target_metric])
        best_hyp = params[HYP][best_idx]
        clf = clone(clf)
        clf.set_params(**{HYP:best_hyp})
        clf.fit(X[train_id],Ytrain[train_id])
        yp = clf.predict(X[test_id])
        for metric in eval_metrics:
            r = metric(Ytest[test_id],yp)
            scores[metric].append(r)
    return scores
```

```python
%%time
noisy_val = test_error(clf,pars,X,yn,yn,y,cv5,CVr,SCORERS,precision_score)
print("Half")
clean_val = test_error(clf,pars,X,yn,y,y,cv5,CVr,SCORERS,precision_score)
```

```python
noisy = pd.DataFrame({k.__name__:v for k,v in noisy_val.items()})
clean = pd.DataFrame({k.__name__:v for k,v in clean_val.items()})
```
