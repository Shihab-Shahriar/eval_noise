```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def convert_time(x):
    if x is None: return None
    try:
        p = datetime.strptime(str(x),"%H:%M:%S")
    except:
        return 0
    return p.hour+p.minute/60
```

```python
df['Deep Work'] = df['Deep Work'].apply(convert_time)
```

```python
plt.figure(figsize=(12,6))
plt.yticks(range(0,11))
plt.xlim(df['Date'][0],df['Date'][len(df)-1])
plt.scatter(df['Date'],df['Deep Work']);
plt.plot(df['Date'],df['Deep Work']);
```
